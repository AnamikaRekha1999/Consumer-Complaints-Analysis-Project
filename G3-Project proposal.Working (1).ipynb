## `2. Detailed description of the dataset:`

The dataset used for this project is the Consumer Complaint Database, which is available on Kaggle. This dataset contains information on consumer complaints received by the Consumer Financial Protection Bureau (CFPB) since 2011. The CFPB is a US government agency responsible for consumer protection in the financial sector.

The dataset includes over 2 million rows, each representing a single consumer complaint. The columns in the dataset provide information on the following aspects of the complaint:

`Product:` The type of product or service the complaint is related to (e.g. credit reporting, debt collection, mortgage, etc.)

`Sub-product:` A more specific description of the product or service (e.g. credit reporting, credit repair services, etc.)

`Issue:` The issue the consumer is facing (e.g. incorrect information on credit report, communication tactics of debt collectors, etc.)

`Sub-issue:` A more specific description of the issue (e.g. information belongs to someone else, debt is not mine, etc.)

`Consumer complaint narrative:` A description of the complaint provided by the consumer
`Company:` The company the complaint is directed towards

`State:` The state in which the consumer resides

`ZIP code:` The zip code in which the consumer resides

`Tags:` Additional information related to the complaint, such as whether the consumer disputed the response or whether the complaint was submitted via the CFPB website or another source

`Consumer consent provided:` Whether the consumer provided consent for the complaint to be shared with regulators or other companies

`Date received:` The date the complaint was received by the CFPB

`Date sent to company:` The date the complaint was sent to the company for a response

`Company response to consumer:` The response provided by the company to the consumer's complaint

`Timely response:` Whether the company provided a timely response to the complaint

`Consumer disputed:` Whether the consumer disputed the company's response

`Complaint ID:` A unique identifier for each complaint


The dataset is a rich source of information on consumer complaints, which can be used to identify patterns and trends in the types of issues consumers face and the companies that receive the most complaints. The dataset aided in the development of models that predicts the category of a new complaint based on its features which was the objective of the project.

# `4.3 Feature Selection:` 



```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Drop rows where the consumer disputed column is missing 
df_raw_data = df_raw_data.dropna(subset=["issue"])

# Encode the target variable
le = LabelEncoder()
df_raw_data["issue"] = le.fit_transform(df_raw_data["issue"])

# Define the features to be used
features = ['product', 'sub_product',  'sub_issue',
            'consumer_complaint_narrative', 'company_public_response', 'company',
            'state', 'zip_code', 'tags', 'consumer_consent_provided',
            'submitted_via', 'date_sent_to_company', 'company_response_to_consumer',
            'timely_response', 'complaint_ID','consumer_disputed']

# Convert categorical features to numerical values
for feature in features:
    df_raw_data[feature] = le.fit_transform(df_raw_data[feature].astype(str))

# Select the top 5 features based on chi-squared test
X = df_raw_data[features]
y = df_raw_data["issue"]
selector = SelectKBest(chi2, k=10)
selector.fit(X, y)

# Print the selected features
selected_features = X.columns[selector.get_support()]
X_selected = X[X.columns[selector.get_support(indices=True)]]

# Print feature importance in a table format
feature_importance = pd.DataFrame({'Features': X_selected.columns, 'Importance': selector.scores_[selector.get_support()]})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

# Print the selected features
print("Selected Features:")
for index, row in feature_importance.iterrows():
    print(f"{row['Features']}: {row['Importance']}")

```

The above code performs feature selection using chi-squared test for the given consumer complaint dataset. It first drops rows with missing values in the "issue" column and encodes the target variable using LabelEncoder. The categorical features are converted to numerical values. It then selects the top 10 features based on chi-squared test and prints the selected features. The features are sorted by importance, with the most important feature at the top. The final output is a list of features and their corresponding importance scores.

# Building the Model

### Random Forest Classifier

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Convert 'text' column to strings
df2['text'] = df2['issue'].astype(str)

# Split the data into input features (X) and target variable (y)
X = df2['text'] 
y = df2['product']  

# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Apply the vectorizer on the input features
X_tfidf = vectorizer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Initialize the Random Forest classifier
rf_classifier = RandomForestClassifier()

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of Random Forest Classifier before calibration :", accuracy)
# Generate and print the classification report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.calibration import CalibratedClassifierCV

# Calibrate the classifier using CalibratedClassifierCV
calibrated_classifier = CalibratedClassifierCV(rf_classifier, method='sigmoid', cv='prefit')
calibrated_classifier.fit(X_train, y_train)

# Make calibrated predictions on the test set
y_calibrated_pred = calibrated_classifier.predict(X_test)

# Evaluate the accuracy of the calibrated classifier
calibrated_accuracy = accuracy_score(y_test, y_calibrated_pred)
print("Random Forest Classifier Accuracy after calibration:", calibrated_accuracy)

# Evaluate the performance of the calibrated classifier
classification_report = classification_report(y_test, y_calibrated_pred)
print(classification_report)
```

To accomplish text classification, the algorithm makes use of a Random Forest classifier. The text column is transformed into strings, and the data is then divided into input features and target variables. The text input is subsequently converted into numerical representations using the TF-IDF vectorizer. The Random Forest classifier is trained using the training data once the data has been divided into training and testing sets. ### SVM, Naive-Bayers & Logistic Regression

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

# Convert 'text' column to strings
df2['text'] = df2['issue'].astype(str)

# Split the data 
X = df2['text']  
y = df2['product'] 

# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Apply the vectorizer on the input features
X_tfidf = vectorizer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Initialize the classifiers
svm_classifier = LinearSVC()
nb_classifier = MultinomialNB()
lr_classifier = LogisticRegression()

# Train the classifiers
svm_classifier.fit(X_train, y_train)
nb_classifier.fit(X_train, y_train)
lr_classifier.fit(X_train, y_train)

# Make predictions on the test set
svm_pred = svm_classifier.predict(X_test)
nb_pred = nb_classifier.predict(X_test)
lr_pred = lr_classifier.predict(X_test)

# Evaluate the accuracy of the classifiers
svm_accuracy = accuracy_score(y_test, svm_pred)
nb_accuracy = accuracy_score(y_test, nb_pred)
lr_accuracy = accuracy_score(y_test, lr_pred)

print("Linear SVM Accuracy:", svm_accuracy)
print("Multinomial Naive Bayes Accuracy:", nb_accuracy)
print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
from sklearn.calibration import CalibratedClassifierCV

# Calibrate Linear SVM classifier
svm_calibrated = CalibratedClassifierCV(svm_classifier, method='sigmoid')
svm_calibrated.fit(X_train, y_train)

# Calibrate Multinomial Naive Bayes classifier
nb_calibrated = CalibratedClassifierCV(nb_classifier, method='sigmoid')
nb_calibrated.fit(X_train, y_train)

# Calibrate Logistic Regression classifier
lr_calibrated = CalibratedClassifierCV(lr_classifier, method='sigmoid')
lr_calibrated.fit(X_train, y_train)

# Make calibrated predictions on the test set
svm_calibrated_pred = svm_calibrated.predict(X_test)
nb_calibrated_pred = nb_calibrated.predict(X_test)
lr_calibrated_pred = lr_calibrated.predict(X_test)

# Evaluate the accuracy of the calibrated classifiers
svm_calibrated_accuracy = accuracy_score(y_test, svm_calibrated_pred)
nb_calibrated_accuracy = accuracy_score(y_test, nb_calibrated_pred)
lr_calibrated_accuracy = accuracy_score(y_test, lr_calibrated_pred)

print("Linear SVM Accuracy after calibration:", svm_calibrated_accuracy)
print("Multinomial Naive Bayes Accuracy after calibration:", nb_calibrated_accuracy)
print("Logistic Regression Accuracy after calibration:", lr_calibrated_accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Initialize the k-NN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the k-NN classifier
knn_classifier.fit(X_train, y_train)

# Make predictions on the test set
knn_pred = knn_classifier.predict(X_test)

# Evaluate the accuracy of the k-NN classifier
knn_accuracy = accuracy_score(y_test, knn_pred)

print("k-Nearest Neighbors Accuracy:", knn_accuracy)

# Calibrate the k-NN classifier
knn_calibrated = CalibratedClassifierCV(knn_classifier, method='sigmoid')
knn_calibrated.fit(X_train, y_train)

# Make calibrated predictions on the test set
knn_calibrated_pred = knn_calibrated.predict(X_test)

# Evaluate the accuracy of the calibrated k-NN classifier
knn_calibrated_accuracy = accuracy_score(y_test, knn_calibrated_pred)

print("k-Nearest Neighbors Accuracy after calibration:", knn_calibrated_accuracy)
```

The programme uses the TF-IDF converted data to train a k-Nearest Neighbors (k-NN) classifier and assesses its accuracy using the test set. It then uses the sigmoid approach to calibrate the k-NN classifier before re-evaluating its accuracy. The accuracy of the k-NN classifier and the accuracy after calibration are the results in a nutshell.

### Ensembling the Classifier Models 

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Initialize the individual classifiers
svm_classifier = SVC(probability=True)  
nb_classifier = MultinomialNB()
lr_classifier = LogisticRegression()
rf_classifier = RandomForestClassifier()
knn_classifier = KNeighborsClassifier()

# Create the ensemble by combining the classifiers
ensemble_classifier = VotingClassifier(
    estimators=[
        ('svm', svm_classifier),
        ('nb', nb_classifier),
        ('lr', lr_classifier),
        ('rf', rf_classifier),
        ('knn', knn_classifier)
    ],
    voting='soft'
)

# Train the ensemble classifier
ensemble_classifier.fit(X_train, y_train)

# Make predictions on the test set
ensemble_pred = ensemble_classifier.predict(X_test)

# Evaluate the accuracy of the ensemble classifier
ensemble_accuracy = accuracy_score(y_test, ensemble_pred)

print("Ensemble Accuracy:", ensemble_accuracy)
```

To get more precise and reliable predictions, ensemble approaches integrate the predictions of various independent classifiers. By utilizing the advantages of several classifiers and minimizing the disadvantages of those same classifiers, they can enhance overall performance.
The provided code combines several classifiers, including Support Vector Machine (SVM), Multinomial Naive Bayes, Logistic Regression, Random Forest, and k-Nearest Neighbors, to construct an ensemble classifier. Soft voting is used by the ensemble classifier to make the final prediction, which considers the probability predicted by each classifier.
 The ensemble classifier is trained using training data, and its accuracy is assessed using test data using the code. The accuracy of the ensemble classifier is the final result in summary.

```python
import nltk
import re
from nltk.corpus import stopwords
import string

nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
stopword=set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = [stemmer.stem(word) for word in text]
    text = " ".join(text)
    return text

df_copy ["consumer_complaint_narrative"] = df_copy ["consumer_complaint_narrative"].apply(clean)
#print(df_dummy )
```

## Priority Column Calculation 

The task of assigning priority values to various concerns based on their existence within particular group IDs is carried out by the code. The priority is determined by counting the instances of each issue within each group ID, which is done by creating a new dataframe that links each group ID to its corresponding issue. The priorities are then scaled using MinMaxScaler to a range of 1 to 5. Priority values are obtained, transformed to integers, and then re-merged into the initial dataframe.

```python
from sklearn.preprocessing import MinMaxScaler

# Create a dataframe for group_id to issue mapping
group_id_df = df2[['group_id', 'issue']].drop_duplicates().reset_index(drop=True)

# Calculate priority based on the count of each issue
group_id_df['priority'] = group_id_df.groupby('group_id')['issue'].transform('count').rank(method='min', ascending=True)

# Scale the priority values to the range of 1 to 5
scaler = MinMaxScaler(feature_range=(1, 5))
group_id_df['priority'] = scaler.fit_transform(group_id_df[['priority']])

# Convert priority values to integers
group_id_df['priority'] = group_id_df['priority'].astype(int)

# Merge the priority column back into the original dataframe
df2 = pd.merge(df2, group_id_df[['group_id', 'priority']], on='group_id', how='left')

# Display the updated dataframe
df2.head()
```
